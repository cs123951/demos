```python
def lcg(x0,n):
    # These choices show the periodicity very well
    # Better choices are a = 16,807 m = 2**31 -1 c = 0
    # Or m = 2**32 a = 1,664,525 c = 1,013,904,223
    # x0 is the seed, use the time of system as the seed
    a = 23
    m = 197
    c = 0
    
    rnd = np.zeros((n))
    
    rnd[0] = np.mod(a*x0 + c,m)
    
    for i in range(1,n):
        rnd[i] = np.mod(a*rnd[i-1]+c,m)
        
    return rnd
```

## CDF变换
对于连续的分布，如果可以计算这个分布的累积分布函数（CDF），就可以通过计算CDF的反函数，结合基础的均匀分布，获得其采样。

假设$F$是一个CDF，$F^{-1}$是其反函数，U是服从均匀分布的变量，则$F^{-1}(U)$服从F的分布。

比如指数分布的概率密度函数是 $f=\lambda e^{-\lambda x}$,($\lambda>0,x>=0$)，其CDF是$y=1-e^{-\lambda x}$，CDF的反函数为$-\frac{ln(1-y)}{\lambda}$

观察指数函数的累积分布函数图，因为CDF的值域为[0,1],这里反函数的作用相当于每次取CDF的y值对应的x值，就可以得到指数分布。


```python
# 指数分布的概率密度函数图
ld = 1
x = np.arange(0, 4, 0.1) 
y = ld*np.exp(-ld*x)
# y = 1 - np.exp(-ld*x)
y_min = ld*np.exp(-ld*4)
y_max = ld*np.exp(ld*4)

# n = 1000  # 产生[y_min,y_max]n个随机数
u = np.random.rand(n)
f_1 = -np.log(1-u)/ld

plt.hist(f_1,normed=1)
plt.plot(x,y)
```

## Box-Muller算法 

证明过程：https://www.cnblogs.com/wlzy/p/7784746.html

Box-Muller变换是通过服从均匀分布的随机变量，来构建服从高斯分布的随机变量的一种方法。

选取两个服从[0,1]上均匀分布的随机变量U1,U2,X,Y满足：

$X=cos(2 \pi U_1) \sqrt{-2lnU_2}$

$Y=sin(2 \pi U_1) \sqrt{-2lnU_2}$

则X与Y服从均值为0，方差为1的高斯分布

证明：

假设有两个独立的正态分布X,Y,由于正态分布的CDF比较难求，所以通过极坐标转换，求CDF

$P(R<r)=-e^{\frac{r^2}{2}}+1$

反函数为 $r=\sqrt{-2ln(1-u)}$

所以$X=R \cdot cos \theta, Y=R \cdot sin \theta$

```python
def boxmuller(n):
    
    x = np.zeros((n,2))
    y = np.zeros((n,2))
    
    for i in range(n):
        [u1, u2] = np.random.rand(2) # [0,1]        
#         y[i] = np.cos(2*pi*u1)*np.sqrt(-2*log(u2))
        y[i,0] = np.cos(2*pi*u1)*np.sqrt(-2*log(u2))
        y[i,1] = np.sin(2*pi*u1)*np.sqrt(-2*log(u2))
#             x2 = x[i,0]*x[i,0]+x[i,1]*x[i,1]
#         y[i,:] = x[i,:] * np.sqrt((-2*np.log(x2))/x2)
    
#     y = np.reshape(y,2*n,1)
    return y

y = boxmuller(1000)
plt.hist(y,normed=1)
x = np.arange(-4,4,0.1)
plt.plot(x,1/np.sqrt(2*np.pi)*np.exp(-0.5*x**2),'k',lw=6)
plt.xlabel('x',fontsize=24)
plt.ylabel('p(x)',fontsize=24)
plt.show()
```

## reject sampling

存在两个问题：（1）合适的M值难找，（2）很难寻找合适的q分布，导致样本点浪费太多

```python
# 这里的4是x的范围，因为x从0到4
def qsample():
    return np.random.rand()*4.

def p(x):
    return 0.3*np.exp(-(x-0.3)**2) + 0.7* np.exp(-(x-2.)**2/0.3) 

def rejection(nsamples):  
    M = 0.8 # 这里的M是因为概率密度并不总是在0和1之间。所以一个M的取值是非常重要的
    samples = np.zeros(nsamples,dtype=float)
    count = 0
    for i in range(nsamples):
        accept = False
        while not accept:
            x = qsample() # 获取采样点
            u = np.random.rand()*M # 判断接收的概率，从均匀分布[0,k*q]中抽样
            if u<p(x):
                accept = True
                samples[i] = x
            else: 
                count += 1
    print(count)  
    return samples

realdata = 0.3*np.exp(-(x-0.3)**2) + 0.7* np.exp(-(x-2.)**2/0.3) 
plt.plot(x,realdata,'k',lw=6)

samples = rejection(10000)

plt.hist(samples,15,density=1)
plt.xlabel('x',fontsize=24)
plt.ylabel('p(x)',fontsize=24)
plt.axis([-0.5,4.5,0,1])
plt.show()
```

## important sampling

```python
def p(x):
    return 0.3*np.exp(-(x-0.3)**2) + 0.7* np.exp(-(x-2.)**2/0.3) 

def q(x):
    return 4.0

def importance_sampling(n):
    
    sample1 = np.zeros(n)
    w = np.zeros(n)
    sample2 = np.zeros(n)
    
    # Sample from q
    sample1 = np.random.rand(n)*4  #从q分布采样n个样本

    # Compute weights
    w = p(sample1)/q(sample1)
    w /= np.sum(w)
    # 权重的概率相加为1，接下来把权重作为概率进行采样。
    # Sample from sample1 according to w
    cumw = np.zeros(len(w))
    cumw[0] = w[0]
    for i in range(1,len(w)):
        cumw[i] = cumw[i-1]+w[i]
    
    u = np.random.rand(n)
    
    index = 0
    for i in range(n):
        indices = np.where(u<cumw[i])
        sample2[index:index+np.size(indices)] = sample1[i]
        index += np.size(indices)
        u[indices]=2
    return sample2

x = np.arange(0,4,0.01)
realdata = 0.3*np.exp(-(x-0.3)**2) + 0.7* np.exp(-(x-2.)**2/0.3) 
plt.plot(x,realdata,'k',lw=6)

nsamples = 1000
samples = importance_sampling(nsamples)
plt.hist(samples,density=1)
plt.xlabel('x',fontsize=24)
plt.ylabel('p(x)',fontsize=24)
plt.axis([-0.5,4.5,0,1])
plt.show()
```

## Metropolis-Hastings算法

考虑到样本之间的转移状态。

但是相邻采样样本之间很有可能关联。解决方法：间隔多次采样再采取一个样本.

```python
def p(x):
    mu1 = 3
    mu2 = 10
    v1 = 10
    v2 = 3
    return 0.3*np.exp(-(x-mu1)**2/v1) + 0.7*np.exp(-(x-mu2)**2/v2)


def q(x, mu, sigma):
    return np.exp(-(x-mu)**2/(sigma**2))

N = 2000
y = np.zeros(N)
mu =5
sigma= 10
y[0] = np.random.normal(mu, sigma)  # 根据q产生一个样本
for i in range(N-1):
    ynew = np.random.normal(mu,sigma)  # 从正态分布产生一个样本, 假设样本互相独立
    alpha = min(1, p(ynew)*q(y[i], mu, sigma)/(p(y[i])*q(ynew, mu, sigma)))  # 计算接受概率
    if np.random.rand() < alpha:
        y[i+1] = ynew
    else:
        y[i+1] = y[i]
plt.figure(1)
x = np.arange(-5,15,0.01)
plt.hist(y,density=True,bins=20)
px = p(x)
plt.plot(x,px*sum(px)/len(x),'k',lw=2)  # *sum(px)/len(x)是缩放，为了让hist和plt重合
plt.show()
```
